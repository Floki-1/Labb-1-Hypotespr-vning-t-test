```python
import numpy as np
# importera biblioteket numpy som np

# skapar en variabel data som innehåller en array av tal
data=np.array([63, 71, 41, 92, 11, 85, 79, 21, 54, 7, 1, 34 ])

# skapar en funktioen mySummary som tar in data som argument och skriver ut olika statistiska mått
def mySummary(data):
    # här använder vi funktion len från python som räknar antalet element i data,
    # och sedan använder vi olika funktioner från numpy för att beräkna medelvärde, median,
    #standardavvikelse, varians, minsta och största värde, samt sortering i stigande och fallande ordning.
    print("antal observationer: ", len(data)) 
    print("Medelvärde: ", np.mean(data))
    print("Median: ", np.median(data))
    print("standardavvikelse: ", np.std(data))
    print("Varians: ", np.var(data))
    print("Minsta värde: ", np.min(data))
    print("Största värde: ", np.max(data))
    print("sortering i stigande ordning: ", np.sort(data))
    # ::-1 används för att vända på sorteringen så att den blir i fallande ordning.
    print("sortering i fallande ordning: ", np.sort(data)[::-1])
    print("interkvartilavstånd: ", np.percentile(data, 75) - np.percentile(data, 25))
    

mySummary(data)
```

```python
## nu ska vi rita histogram, vi impoterar biblioteket matplotlib.pyplot som plt

import matplotlib.pyplot as plt

# vi skapar en egen funktion myhistogram som ska visa ett histogram av data. tar data som arguent. 
def myhistogram(data):
    # vi använder funktionen hist från matplotlib för att rita histogrammet.
    # bins bestämmer hur många staplar det ska vara i histogrammet,
    #  color bestämmer färgen på staplarna och edgecolor bestämmer färgen på kanterna av staplarna.

    plt.hist(data, bins=10,color="blue", edgecolor="black")
    # funktionen title sätter en titel på diagrammet, xlabel och ylabel sätter namn på x- och y-axeln.
    plt.title("Histogram")
    plt.xlabel("Värden") 
    plt.ylabel("antal observationer")

# anropar funktionen myhistogram och skickar in data som argument för att visa histogrammet.
myhistogram(data)    



# Mått beskrivning
# det vi ser ....
```

```python
## nu ska vi rita stapeldiagram.
def mystapeldiagram(data):
    # vi använder funktionen bar från matplotlib för att rita stapeldiagrammet.
    # x är en array som innehåller indexen för varje stapel, och y är data som innehåller höjden på varje stapel.
    x = np.arange(len(data))
    y = data
    plt.bar(x, y, color="green", edgecolor="black")
    plt.title("Stapeldiagram")
    plt.xlabel("Index")
    plt.ylabel("Värden")

mystapeldiagram(data)   
```
