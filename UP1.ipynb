```python
import numpy as np

data=np.array([63, 71, 41, 92, 11, 85, 79, 21, 54, 7, 1, 34 ])

def mySummary(data):
    print("Medelvärde: ", np.mean(data))
    print("Median: ", np.median(data))
    print("standardavvikelse: ", np.std(data))
    print("Varians: ", np.var(data))
    print("Minsta värde: ", np.min(data))
print("Största värde: ", np.max(data))
print("sortering i stigande ordning: ", np.sort(data))
print("sortering i fallande ordning: ", np.sort(data)[::-1])

print("interkvartilavstånd ???") 

mySummary(data)
```
